```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import statsmodels.api as sm 
import statsmodels.formula.api as sf
import numpy as np
```

```python
df = pd.read_csv('data/titanik_full_data.csv', sep = '\t')
df.head()
```

```python
mod = sf.glm('Survived ~ C(Pclass) + C(Sex) + Age',
            data = df,
            family = sm.families.Binomial()).fit()

mod.summary()
```

- __Псевдо R-квадрат равен 0.4421 => наша модель объясняет 44.21% изменчивости переменной выживаемости__
- __p-значения для всех независимых переменных меньше 0.001 => вероятность того, что мы получили данные коэффициенты случайно, меньше 0.001__
- __Первый класс и женский пол в данной модели являются референтными переменными__
- __Выживаемость пассажиров первого класса больше, чем второго, и значительно больше, чем третьего__
- __Выживаемость женщин значительно больше выживаемости мужчин__
- __Выживаемость обратна пропорциональна возрасту, но с очень небольшим коэффициентом__
