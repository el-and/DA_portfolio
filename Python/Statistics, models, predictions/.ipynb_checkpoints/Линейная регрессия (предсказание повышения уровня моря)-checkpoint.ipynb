```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from scipy.stats import linregress
```

Наша __цель__ - предсказать повышение уровня моря к 2050 году, используя данные о повышения уровня моря за период с 1880 по 2015 год _(CSIRO (Commonwealth Scientific and Industrial Research Organisation). 2017 update to data originally published in: Church, J.A., and N.J. White. 2011. Sea-level rise from the late 19th to the early 21st century. Surv. Geophys. 32:585–602. Accessed September 2017; NOAA (National Oceanic and Atmospheric Administration). 2022. Laboratory for Satellite Altimetry: Sea level rise. Accessed March 2022)_.

Мы используем данные CSIRO, поскольку в них нет пустых значений.

```python
df = pd.read_csv('data/epa-sea-level.csv')
df.head()
```

```python
df.isna().any()
```

```python
fig, ax = plt.subplots(figsize = (9, 6))
ax.scatter('Year', 'CSIRO Adjusted Sea Level', data = df);

res1 = linregress(df['Year'], df['CSIRO Adjusted Sea Level'])
x1 = range(df['Year'].min(), 2051)
y1 = x1 * res1.slope + res1.intercept
ax.plot(x1, y1, 'r')

# Поскольку в последние 20 лет скорость, с которой уровень моря повышается, заметно увеличилась, 
# построим ещё одну регрессионную модель, уже только на данных с 2000 года
df_rec = df[df['Year'] >= 2000]
res2 = linregress(df_rec['Year'], df_rec['CSIRO Adjusted Sea Level'])
x2  = range(2000, 2051)
y2 = x2 * res2.slope + res2.intercept
ax.plot(x2, y2, 'g')

ax.set_xlabel('Year')
ax.set_ylabel('Sea Level (inches)')
ax.set_title('Rise in Sea Level')

plt.text(2050, 2050*res1.slope + res1.intercept, '{:.2f}'.format(2050*res1.slope + res1.intercept), 
         ha = 'center', 
         va = 'bottom')
plt.text(2050, 2050*res2.slope + res2.intercept, '{:.2f}'.format(2050*res2.slope + res2.intercept),  
         ha = 'center', 
         va = 'bottom')

plt.show()
```

- __Значения r-value для обеих моделей > 0.97 => наша модель объясняет 97% изменений зависимой переменной__
- __Значения p-value для обеих моделей меньше 1е-08 => вероятность того, что коэффициент на самом деле равен 0, меньше 1е-08__
- __Модель, построенная на данных с 1880 года, предсказывает повышение уровня моря на 10.18 дюймов к 2050 году__
- __Модель, построенная на данных с 2000 года, предсказывает повышение уровня моря на 15.38 дюймов к 2050 году__
