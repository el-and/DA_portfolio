```python
import datetime
from datetime import datetime as dt
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

pd.set_option('display.max_columns', None)
```

## Подготовка данных 

Мы используем данные о бронированиях в двух отелях (курортном и городском), расположенных в Португалии. (Antonio, N., de Almeida, A., & Nunes, L. (2019). Hotel booking demand datasets. *Data in Brief, 22*, 41–49. https://doi.org/10.1016/j.dib.2018.11.126)

Наша __цель__ - обработать и визуализировать данные, чтобы понять, какие факторы могут влиять на отмену бронирования.

Мы начинаем со слияния датасетов из двух отелей в один. Затем мы модифицируем данные, чтобы упростить работу с ними:
- убираем пустые символы в начале и конце всех строчек
- превращаем тип данных ReservationStatusDate в datetime
- превращаем тип данных IsCanceled и IsRepeatedGuest в bool
- создаём новый datetime-столбец для даты прибытия
- заменяем строку 'NULL' настоящими пустыми значениями
- проверяем наличие пустых значений и заменяем их тем, что имеет смысл (для 'Country', 'Agent' и 'Company' мы используем 'Unknown', для 'Children' - 0)

```python
res_df = pd.read_csv('data/H1.csv')
city_df = pd.read_csv('data/H2.csv')
city_df['Hotel'] = 'City'
res_df['Hotel'] = 'Resort'
df = pd.concat([city_df, res_df], axis = 0, ignore_index = True)
df.head()
```

```python
df.info()
```

```python
df[df.select_dtypes('object').columns] = df[df.select_dtypes('object').columns].apply(lambda x: x.str.strip())
df['ReservationStatusDate'] = pd.to_datetime(df['ReservationStatusDate'])
df['IsCanceled'] = df['IsCanceled'].astype('bool')
df['IsRepeatedGuest'] = df['IsRepeatedGuest'].astype('bool')

df['ArrivalDate'] = pd.to_datetime(dict(
                        year = df['ArrivalDateYear'],
                        month = pd.Series(map(lambda x: dt.strptime(x, '%B').month, df['ArrivalDateMonth'])),
                        day = df['ArrivalDateDayOfMonth']))

df = df.replace('NULL', np.nan)
```

```python
df.isnull().any()
```

```python
[' - '.join((str(i), str(df[i].isnull().sum()))) for i in ['Children', 'Country', 'Agent', 'Company']]
```

```python
df['Children'] = df['Children'].fillna(0)
df = df.fillna('Unknown')
```

NB: в Company большая часть значений пустые, что делает эту переменную плохо пригодной для анализа

## EDA

```python
df.describe(percentiles = [0.25, 0.5, 0.75, 0.99])
```

Первое, что мы замечаем - очевидные выбросы в некоторых численных переменных Adults, Children, Babies, StaysInWeekNights, etc). Возможно, некоторое количество отмен вызвано ошибками во время бронирования (10 младенцев и 55 взрослых могут быть группой, а могут быть опечаткой). Выбросим значения, лежащие ниже 0.01- и выше 0.99-квантили и посмотрим, как это влияет на долю отменённых бронирований.

```python
vc100 = df['IsCanceled'].value_counts()

vc99 = df[(df.Adults <= df.Adults.quantile(.99)) &
  (df.Children <= df.Children.quantile(.99)) &
  (df.Babies <= df.Babies.quantile(.99)) &
  (df.StaysInWeekendNights <= df.StaysInWeekendNights.quantile(.99)) &
  (df.StaysInWeekNights <= df.StaysInWeekNights.quantile(.99)) &
  (df.ADR <= df.ADR.quantile(.99))
  ]['IsCanceled'].value_counts()

vc01 = df[(df.Adults > df.Adults.quantile(.99)) |
  (df.Children > df.Children.quantile(.99)) |
  (df.Babies > df.Babies.quantile(.99)) |
  (df.StaysInWeekendNights > df.StaysInWeekendNights.quantile(.99)) |
  (df.StaysInWeekNights > df.StaysInWeekNights.quantile(.99)) |
  (df.ADR > df.ADR.quantile(.99))
  ]['IsCanceled'].value_counts()

print('% of cancelled in original df: {:.2f}% \
\n% of cancelled without outliers: {:.2f}% \
\n% of cancelled within outliers: {:.2f}%'.format(vc100[0] / vc100.sum()*100, 
                                                  vc99[0] / vc99.sum()*100, 
                                                  vc01[0] / vc01.sum()*100))
```

__Без выбросов процент отмены бронирований снижается примерно на 0.1%. Мы смело можем предположить, что они мало влияют на общее число отмен. Более того, процент отмен среди выбросов тоже не повышается значимо, что означает, что это не ошибки, которые приводят к отмене бронирований, и мы можем оставить их в нашем датафрейме.__

### Время ожидания / задержки:

Время ожидания (waiting time) - это количество дней между появлением бронирования в листе ожидания и подтверждением его клиенту.

Время задержки (Lead time) - это количество дней между появлением бронирования в базе данных отеля и датой прибытия.

Из-за выбросов мы устанавливаем ограничения на ось х.

```python
fig, (ax1, ax2) = plt.subplots(2, figsize = (12, 9))

sns.kdeplot(data = df,
            x = 'DaysInWaitingList',
            hue = 'IsCanceled',
           fill = True,
           ax = ax1)

sns.kdeplot(data = df,
            x = 'LeadTime',
            hue = 'IsCanceled',
           fill = True,
           ax = ax2)


ax1.set_xlim(0, df['DaysInWaitingList'].quantile(.99))
ax2.set_xlim(0, df['LeadTime'].quantile(.99))
plt.show()
```

- __В обоих случаях распределение плотности для отменённых бронирований шире, чем для не отменённых__
- __Среди бронирований с временем ожидания больше 4 дней отменённых больше, чем активных__
- __То же самое справедливо для бронирований с временем задержки больше 240 дней__
